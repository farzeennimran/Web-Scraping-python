```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chromium.webdriver import ChromiumDriver
from selenium.webdriver.common.by import By
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
options.add_argument('no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
driver = webdriver.Chrome(options)
```

```python
urlPath = 'https://www.youtube.com/@NASA/videos'
driver.get(urlPath)
```

```python
videos = driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-item-renderer")

for video in videos:
    titles = driver.find_elements(By.XPATH,'.//*[@id="video-title"]')
    views = driver.find_elements(By.XPATH,'.//*[@id="metadata-line"]/span[1]')
    links = driver.find_elements(By.XPATH, './/*[@id="video-title-link"]')
    dates = driver.find_elements(By.XPATH, './/*[@id="metadata-line"]/span[2]')
    images = driver.find_elements(By.XPATH, './/*[@id="thumbnail"]')
```

```python
# Find and extract image URLs
image_elements = driver.find_elements(By.TAG_NAME, 'img')
image_urls = [img.get_attribute('src') for img in image_elements if img.get_attribute('src')]
```

```python
titlesData = []
for title in titles:
    titlesData.append(title.text)
    # print(title.text)

viewsData = []
for view in views:
    viewsData.append(view.text)
    # print(view.text)

linksData = []
for link in links:
  linksData.append(link.text)

datesData = []
for date in dates:
  datesData.append(date.text)

imagesData = []
for image in images:
  imagesData.append(image.text)

import pandas as pd
df = pd.DataFrame({'title': titlesData, 'views': viewsData, 'links' : linksData, 'dates' : datesData})
df.to_csv('youtube.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
URL = "https://www.dawnnews.tv/tech"
page = requests.get(URL)
print(page.text)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
print(soup)
```

```python
story_title = soup.find('h2', class_='story__title      text-6.5  text-colorsDawn-green hover:text-colorsDawn-red-darkest font-bold leading-tight    pt-0.75  pb-2  ')
story_time = soup.find('span', class_='timestamp--time  timeago')
media_item = soup.find('h2', class_='media__item          ')
story_link = soup.find('a', class_='story__link  ')
story_excerpt = soup.find('div', class_='story__excerpt')
```

```python
import pandas as pd
csvfile = pd.DataFrame({'story Title': story_title, 'story Time': story_time, 'media Item' : media_item, 'story Link' : story_link, 'story Excerpt' : story_excerpt}, index=[0])
csvfile.to_csv('youtube.csv')
```
